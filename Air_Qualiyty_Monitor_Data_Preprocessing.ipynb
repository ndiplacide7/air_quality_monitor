```python
import pandas as pd
```

```python
air_quality_df = pd.read_json("https://www.data.act.gov.au/resource/94a5-zqnn.json")
air_quality_df
```

```python
air_quality_df[air_quality_df.isnull().any(axis=1)]
```

```python
air_quality_df.columns
```

```python
# Count missing values in each column
missing_values = air_quality_df.isnull().sum()

# Display columns with missing values
columns_with_missing = missing_values[missing_values > 0]
print(columns_with_missing)
```

```python
our_dbf = air_quality_df[['Name','NO2', 'O3_4hr', 'CO', 'PM2.5', 'PM10','PM2.5', 'AQI_CO', 'AQI_NO2', 'AQI_O3_4hr','AQI_PM10', 'AQI_PM2.5',
                          
                          ]]

our_dbf
```

```python
# List of columns to clean
columns_to_clean = ['NO2', 'O3_4hr', 'CO', 'PM2.5', 'PM10', 'AQI_CO', 'AQI_NO2', 'AQI_O3_4hr', 'AQI_PM10', 'AQI_PM2.5']

# Replace NaN values with the mean for each column
for column in columns_to_clean:
    our_dbf[column] = our_dbf[column].fillna(our_dbf[column].mean())

our_dbf
```

```python
print(our_dbf.isnull().sum())
```

```python
# Calculate the overall AQI as the maximum of the individual AQIs
our_dbf['AQI_overall'] = our_dbf[['AQI_NO2', 'AQI_O3_4hr', 'AQI_CO', 'AQI_PM10', 'AQI_PM2.5']].max(axis=1)

# Now the 'AQI_overall' column contains the overall AQI for each row
our_dbf
```

```python
our_dbf
```

```python
# Get unique categories in the 'Name' column
unique_categories = our_dbf['Name'].unique()

# Print the unique categories
print(unique_categories)
```

```python
# Group by the 'Name' column and calculate the mean of 'AQI_overall' for each category
average_aqi_by_name = our_dbf.groupby('Name')['AQI_overall'].mean()

# Print the result
print(average_aqi_by_name)
```
